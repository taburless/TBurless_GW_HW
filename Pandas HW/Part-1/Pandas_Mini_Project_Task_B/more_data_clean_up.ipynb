```python
# Dependencies
import pandas as pd
```

```python
# load CSV
csv_path = "Resources/2016-FCC-New-Coders-Survey-Data.csv"
```

```python
# Read with pandas--low_memory required to suppress errors about mixed data types
more_clean_data = pd.read_csv(csv_path)
more_clean_data.head()
```

```python
# Take only columns 0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
more_reduced_clean = more_clean_data.iloc[:,[ 0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
more_reduced_clean.head()
```

```python
# Change reading 0 and 1 to No and Yes, respectively
more_reduced_clean = more_reduced_clean.replace({0.0: "No", 1.0: "Yes"})
more_reduced_clean.head()
```

```python
# Extract rows for only those who attended a bootcamp
more_attended = more_reduced_clean.loc[more_reduced_clean["AttendedBootcamp"] == "Yes"]
more_attended.count()
```

```python
more_rec = more_reduced_clean.loc[more_reduced_clean["BootcampRecommend"] == "Yes"]
more_rec.count()
```

```python
# Create DataFrame of the different boot camps that had a significant number of attendees
#group_bootcamp = more_attended.groupby('BootcampName')
#group_bootcamp
#count_bootcamp = group_bootcamp['BootcampName'].count()
#count_bootcamp
att_bootcamp = more_attended.groupby(["BootcampName"]).count()["AttendedBootcamp"].rename("Attended the Bootcamp")

bootcamp_data_attend = pd.DataFrame({"Attended the Bootcamp": att_bootcamp })
bootcamp_data_attend.head(10)
```

```python
# Count how many attendees of each bootcamp would recommend the bootcamp
rec_bootcamp = more_rec.groupby(["BootcampName"]).count()["BootcampRecommend"].rename("Recommends the Bootcamp")

bootcamp_data_recommend = pd.DataFrame({"Recommends the Bootcamp": rec_bootcamp })
bootcamp_data_recommend.head(10)
```

```python
# Merge the two created data frames on the name of tbe bootcamp
merge_table = pd.merge(bootcamp_data_attend, bootcamp_data_recommend, on="BootcampName")
merge_table.head(10)
```

```python
# Calculate percentage of eac bootcamp's students who are recommenders
merge_table['Result'] = (merge_table['Recommends the Bootcamp']/merge_table['Attended the Bootcamp']).map("{0:,.2f}%".format)
merge_table.head(10)
```

```python
# Export to excel and remove index
export_excel = merge_table.to_excel (r'Resources/more_data_finished.xlsx', index = None, header=True)
```
