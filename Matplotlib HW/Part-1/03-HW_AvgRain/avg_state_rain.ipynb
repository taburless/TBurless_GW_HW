```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load in csv
avg_rain_df = pd.read_csv("Resources/avg_rain_state.csv")
avg_rain_df.head()
```

```python
# Set x axis and tick locations
x_axis = np.arange(len(avg_rain_df))
tick_locations = [value+0.4 for value in x_axis]
```

```python
# Create a list indicating where to write x labels and set figure size to adjust for space
plt.figure(figsize=(10,3))
plt.bar(x_axis, avg_rain_df["Inches"], color='y', alpha=0.75, align="edge")
plt.xticks(tick_locations, avg_rain_df["State"], rotation="vertical")
```

```python
# Set x and y limits
plt.xlim(0, len(x_axis))
plt.ylim(0, max(avg_rain_df["Inches"]+10))
```

```python
# Set a Title and labels
plt.title("Average Rain Fall per State")
plt.ylabel("Rainfall(Inches)")
plt.xlabel("States")
```

```python
# Save our graph and show the grap
plt.tight_layout()
plt.savefig("Images/avg_state_rain.png")
plt.show()
```
