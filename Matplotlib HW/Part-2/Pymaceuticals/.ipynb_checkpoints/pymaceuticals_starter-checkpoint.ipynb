```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
#import scipy from stats

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_data = pd.read_csv(mouse_drug_data_to_load)
clinical_trial_data = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
clinicals_combined = pd.merge(clinical_trial_data, mouse_drug_data, how="left", on=["Mouse ID", "Mouse ID"])

# Display the data table for preview
clinicals_combined.head()

```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
grouped_tumor_avg = clinicals_combined.groupby(["Drug", "Timepoint"])["Tumor Volume (mm3)"].mean()
# Convert to DataFrame
grouped_tumor_summary = pd.DataFrame(grouped_tumor_avg).reset_index()
#grouped_tumor_summary.rename(columns={'Tumor Volume (mm3)':'Mean Tumor Volume'},inplace = True)
# Preview DataFrame
grouped_tumor_summary.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
grouped_tumor_se = clinicals_combined.groupby(["Drug", "Timepoint"])["Tumor Volume (mm3)"].sem()
# Convert to DataFrame
tumor_se_summary = pd.DataFrame(grouped_tumor_se).reset_index()
# Preview DataFrame
tumor_se_summary.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
reformat_tumor_mean = grouped_tumor_summary.pivot(index = "Timepoint", columns = "Drug", values = "Tumor Volume (mm3)")
reformat_tumor_se = tumor_se_summary.pivot(index = "Timepoint", columns = "Drug", values = "Tumor Volume (mm3)")
# Preview that Reformatting worked
reformat_tumor_mean.head()
```

```python
# Generate the Plot (with Error Bars)
capomulin_mean = reformat_tumor_mean.loc[:, ["Capomulin"]]
capomulin_mean = capomulin_mean["Capomulin"]

capomulin_se = reformat_tumor_se.loc[:, ["Capomulin"]]
capomulin_se = capomulin_se["Capomulin"]

ketapril_mean = reformat_tumor_mean.loc[:, ["Ketapril"]]
ketapril_mean = ketapril_mean["Ketapril"]

ketapril_se = reformat_tumor_se.loc[:, ["Ketapril"]]
ketapril_se = ketapril_se["Ketapril"]

infubinol_mean = reformat_tumor_mean.loc[:, ["Infubinol"]]
infubinol_mean = infubinol_mean["Infubinol"]

infubinol_se = reformat_tumor_se.loc[:, ["Infubinol"]]
infubinol_se = infubinol_se["Infubinol"]

placebo_mean = reformat_tumor_mean.loc[:, ["Placebo"]]
placebo_mean = placebo_mean["Placebo"]

placebo_se = reformat_tumor_se.loc[:, ["Placebo"]]
placebo_se = placebo_se["Placebo"]
#fig, ax = plt.subplots()

x_axis = reformat_tumor_mean.index

plt.errorbar(x_axis, capomulin_mean, capomulin_se, fmt="o--", label = "Capomulin")
plt.errorbar(x_axis, ketapril_mean, ketapril_se, fmt="^--", label = "Ketapril")
plt.errorbar(x_axis, infubinol_mean, infubinol_se, fmt="s--", label = "Infubinol")
plt.errorbar(x_axis, placebo_mean, placebo_se, fmt="D--", label = "Placebo")

plt.legend(loc="best")
plt.title("Tumor Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.grid(True)


# Save the Figure

plt.savefig("../Images/TumorVolume_Scatter.png")
```

```python
# Show the Figure
plt.show()
```

![Tumor Response to Treatment](../Images/treatment.png)

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
metastatic_mean = clinicals_combined.groupby(["Drug", "Timepoint"])["Metastatic Sites"].mean()
# Convert to DataFrame
metastatic_mean_summary = pd.DataFrame(metastatic_mean).reset_index()
# Preview DataFrame
metastatic_mean_summary.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
metastatic_se = clinicals_combined.groupby(["Drug", "Timepoint"])["Metastatic Sites"].sem()
# Convert to DataFrame
metastatic_se_summary = pd.DataFrame(metastatic_se).reset_index()
# Preview DataFrame
metastatic_se_summary.head()
```

```python
# Generate the Plot (with Error Bars)
capomulin_meta_mean = reformat_metastatic_mean.loc[:, ["Capomulin"]]
capomulin_meta_mean = capomulin_meta_mean["Capomulin"]

capomulin_meta_se = reformat_metastatic_se.loc[:, ["Capomulin"]]
capomulin_meta_se = capomulin_meta_se["Capomulin"]

ketapril_meta_mean = reformat_metastatic_mean.loc[:, ["Ketapril"]]
ketapril_meta_mean = ketapril_meta_mean["Ketapril"]

ketapril_meta_se = reformat_metastatic_se.loc[:, ["Ketapril"]]
ketapril_meta_se = ketapril_meta_se["Ketapril"]

infubinol_meta_mean = reformat_metastatic_mean.loc[:, ["Infubinol"]]
infubinol_meta_mean = infubinol_meta_mean["Infubinol"]

infubinol_meta_se = reformat_metastatic_se.loc[:, ["Infubinol"]]
infubinol_meta_se = infubinol_meta_se["Infubinol"]

placebo_meta_mean = reformat_metastatic_mean.loc[:, ["Placebo"]]
placebo_meta_mean = placebo_meta_mean["Placebo"]

placebo_meta_se = reformat_metastatic_se.loc[:, ["Placebo"]]
placebo_meta_se = placebo_meta_se["Placebo"]
#fig, ax = plt.subplots()

x_axis = reformat_tumor_mean.index

plt.errorbar(x_axis, capomulin_meta_mean, capomulin_meta_se, fmt="o--", label = "Capomulin")
plt.errorbar(x_axis, ketapril_meta_mean, ketapril_meta_se, fmt="^--", label = "Ketapril")
plt.errorbar(x_axis, infubinol_meta_mean, infubinol_meta_se, fmt="s--", label= "Infubinol")
plt.errorbar(x_axis, placebo_meta_mean, placebo_meta_se, fmt="D--", label = "Placebo")

plt.legend(loc="best")
plt.title("Metastatic Spread During Treatment")
plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Met. Sites")
plt.grid(True)
# Save the Figure
plt.savefig("../Images/Metastatic_Scatter.png")
# Show the Figure
plt.show
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
#final_mean = reformat_tumor_mean.iloc[9, :]
#first_mean = reformat_tumor_mean.iloc[0, :]
#vol_percent_change = ((final_mean - first_mean) / first_mean) * 100
vol_percent_change = reformat_tumor_mean.pct_change(periods = 9).dropna() *100
vol_percent_change = vol_percent_change.reset_index()
# Display the data to confirm
vol_percent_change
```

```python
# Store all Relevant Percent Changes into a Tuple
#percent_change_capomulin = vol_percent_change["Capomulin"]
#percent_change_ketapril = vol_percent_change["Ketapril"]
#percent_change_infubinol = vol_percent_change["Infubinol"]
#percent_change_placebo = vol_percent_change["Placebo"]

tuple_percent_changes = (vol_percent_change["Capomulin"], 
                         vol_percent_change["Ketapril"], 
                         vol_percent_change["Infubinol"], 
                         vol_percent_change["Placebo"])

# Splice the data between passing and failing drugs

fig, ax = plt.subplots()

ind = np.arange(len(tuple_percent_changes))

width = 1       # the width of the bars: can also be len(x) sequence

rectsPass = ax.bar(ind[0], tuple_percent_changes[0], width, color="green")
rectsFail_x = ax.bar(ind[1], tuple_percent_changes[1], width, color="red")
rectsFail_y = ax.bar(ind[2], tuple_percent_changes[2], width, color="red")
rectsFail_z = ax.bar(ind[3], tuple_percent_changes[3], width, color="red")

# Orient widths. Add labels, tick marks, etc. 
ax.set_ylabel("% Tumor Volume Change")
ax.set_title("Tumor Change Over 45 Day Treatment")
ax.set_xticks(ind + 0.5)
ax.set_xticklabels(('Capomulin', 'Ketapril', 'Infubinol', 'Placebo'))
ax.set_autoscaley_on(False)
ax.set_ylim([-30, 70])
ax.grid(True)

# Use functions to label the percentages of changes
def autolabelFail(rects):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 3, "%d%%" % int(height),
               ha = 'center', va = 'bottom', color = 'white')

def autolabelPass(rects):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., -8, "-%d%%" % int(height),
               ha = 'center', va = 'bottom', color = 'white')

# Call functions to implement the function calls
autolabelPass(rectsPass)
autolabelFail(rectsFail_x)
autolabelFail(rectsFail_y)
autolabelFail(rectsFail_z)

# Save the Figure

# Show the Figure
#fig.show()
ind
rectsPass
rectsFail
```
